## Using spaCy:
The loop below cleans the body text slightly and uses spaCy to lemmatize the text and remove stopwords, punctuation, and spaces

```python
#now, we want to use spaCy to remove stopwords, punctuation, spaces, etc
#then, we'll be able to use bag of words on what remains

cleaner = []
n = len(questions)
for j in range(0,n-1):
    no_newline0 = re.sub(r'\n',' ',questions['body_text'][j])
    no_newline1 = re.sub(r';',' ',no_newline0)
    testdoc0 = nlp(no_newline1)
    test_nostop0_0 = [token.lemma_ for token in testdoc0 if not token.is_stop | token.is_punct | token.is_space]
    cleaner.append(test_nostop0_0)
    if (j % 10000 -1 == -1):
        print(j)
```
